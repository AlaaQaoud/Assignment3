```python
import pandas as pd
data=[
    [50,30,140,180,'Yes'],
    [40,25,130,160,'No'],
    [45,28,150,170,'Yes'],
    [35,22,120,140,'No'],
    [60,35,160,200,'Yes']
]
columns=[
    'Age','BMI','Blood Pressure','Glucose Level','Diabetic (Yes/No)'
]
df=pd.DataFrame(data,columns=columns)
df
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

d = {'Yes': 1, 'No': 0}
df['Diabetic (Yes/No)'] = df['Diabetic (Yes/No)'].map(d)

features = ['Age','BMI','Blood Pressure','Glucose Level']

X = df[features]
y = df['Diabetic (Yes/No)']

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features)
```
